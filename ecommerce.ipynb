E-commerce Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv("Order Details.csv")
data
```

```python
data.head()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
total_orders = data['Order ID'].nunique()
print("Total Orders:", total_orders)
```

```python
#total revenue
total_revenue = data['Amount'].sum()
print("Total Revenue:", total_revenue)
```

```python
#Average Order Amount
average_order_amount = data['Amount'].mean()
print("Average Order Amount:", average_order_amount)
```

```python
#Popular Category
popular_categories = data['Category'].value_counts().head(5)
print("Most Popular Categories:")
print(popular_categories)
```

```python
popular_subcategories = data['Sub-Category'].value_counts().head(5)
print("Most Popular Sub-Categories:")
print(popular_subcategories)
```

```python
#Histogram for order amount
plt.figure(figsize=(10, 6))
plt.hist(data['Amount'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Order Amount')
plt.ylabel('Frequency')
plt.title('Distribution of Order Amounts')
plt.show()
```

```python
#Bar plot for popular categories
plt.figure(figsize=(10, 6))
popular_categories.plot(kind='bar', color='green')
plt.xlabel('Category')
plt.ylabel('Number of Orders')
plt.title('Most Popular Categories')
plt.show()
```

```python
#Bar plot for popular subcategories
plt.figure(figsize=(10, 6))
popular_subcategories.plot(kind='bar', color='red')
plt.xlabel('Sub_Category')
plt.ylabel('Number of Orders')
plt.title('Most Popular Categories')
plt.show()
```

```python
# Pie chart for distribution of subcategories
subcategory_distribution = data['Sub-Category'].value_counts()
plt.figure(figsize=(10, 6))
plt.pie(subcategory_distribution, labels=subcategory_distribution.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Distribution of Subcategories')
plt.show()
```
